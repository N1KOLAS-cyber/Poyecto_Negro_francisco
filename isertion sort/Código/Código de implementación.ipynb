```python
import numpy as np
import matplotlib.pyplot as plt
import time

# Implementaci√≥n del algoritmo de ordenamiento por inserci√≥n
def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and arr[j] > key:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key

# Funci√≥n para medir el tiempo de ejecuci√≥n del algoritmo
def measure_time(n, case="average"):
    np.random.seed(42)  # Para reproducibilidad
    if case == "best":
        arr = list(range(n))  # Lista ya ordenada (mejor caso)
    elif case == "worst":
        arr = list(range(n, 0, -1))  # Lista en orden inverso (peor caso)
    else:
        arr = list(np.random.randint(0, 1000, n))  # Caso promedio
    
    start_time = time.time()
    insertion_sort(arr)
    return time.time() - start_time

# Valores de n para las pruebas
sizes = np.linspace(100, 2000, 20, dtype=int)

# Medir tiempos para los tres casos
best_times = [measure_time(n, "best") for n in sizes]
worst_times = [measure_time(n, "worst") for n in sizes]
average_times = [measure_time(n, "average") for n in sizes]

# Graficar los resultados
plt.figure(figsize=(10, 6))
plt.plot(sizes, best_times, label="Mejor caso (O(n))", marker='o')
plt.plot(sizes, worst_times, label="Peor caso (O(n¬≤))", marker='s')
plt.plot(sizes, average_times, label="Caso promedio (O(n¬≤))", marker='^')

plt.xlabel("Tama√±o de la lista (n)")
plt.ylabel("Tiempo de ejecuci√≥n (segundos)")
plt.title("An√°lisis de complejidad del Ordenamiento por Inserci√≥n")
plt.legend()
plt.grid()
plt.show()
```
